```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['species'] = iris.target
```

```python
print(iris.target_names)
```

```python
df.species = df.species.map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
df.head()
```

```python
inputs = df.drop(columns="species")
target = iris.target
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.25)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.predict_proba(X_test)[1:11]
```

```python
#the three desicion trees are classifying data
```

```python
model.score(X_test,y_test)
```

```python
predicted = X_test
predicted['predicted_Val'] = model.predict(X_test)
predicted['Actual_Val'] = y_test
```

```python
predicted
```

```python
predicted.predicted_Val = predicted.predicted_Val.map({0:'setosa',1:'versicolor',2:'virginica'})
predicted.Actual_Val = predicted.Actual_Val.map({0:'setosa',1:'versicolor',2:'virginica'})
predicted.head(10)
```
